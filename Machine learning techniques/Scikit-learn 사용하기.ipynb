이미 누군가 만들어놓은 라이브러리를 활용해보자. sklearn이라는 라이브러리는 여러가지 기능을 제공한다.
-  dataset. 오늘은 Boston의 집값 데이터셋(load_boston)을 이용하기로 한다.
- train set과 test set을 분리해주는 함수
- Linear regression을 이미 구현해놓은 함수
- MSE를 구현해놓은 함수


```
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
```

boston_dataset이라는 변수에 데이터셋을 저장한다.
load_boston의 type을 확인해보자

```
boston_dataset = load_boston()
```

```
print(type(load_boston))
print(type(load_boston()))
print(type(boston_dataset))
```

boston_dataset의 자세히 살펴보자. DESCR는 describe의 약자

보스턴 집 506개에 대한 데이터가 있다.
입력 변수는 총 13개가 있는데, CRIM(범죄율), RM(방 개수), RAD(고속도로 접근성) 등이 있다.

이 데이터들로 14번째 target attribute인 MED(집값)을 알아내는 것이 목표이다.



```
print(boston_dataset.DESCR)
```

dataset의 feature 이름 알아내기

```
boston_dataset.feature_names
```

dataset의 모양. 506 x 13 matrix임을 알 수 있다.

```
boston_dataset.data.shape
```

```
boston_dataset.target.shape
```

원하는 column만 뽑아내기 위해 pandas의 DataFrame으로 변환한다. 

주의할 것은 boston_dataset은 type이 함수라는 것. 따라서 boston_dataset.data로 넘파이 어레이만 뽑아내야 한다.

```
type(boston_dataset.data)
```

```
x = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
x
```

편의를 위해 변수를 하나(AGE)로 제한한다.

```
x = x[['AGE']]
```

```
y = pd.DataFrame(boston_dataset.target, columns=['MEDV'])
```

train_test_split 함수를 이용해 train set과 test set을 나눌 수 있다.

test_size = 0.2를 넘기면 전체 데이터셋의 20 %가 test set으로 분리된다.(train : test = 4 : 1)
분리된 데이터셋은 dataframe 타입


```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)
```

```
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

print(type(x_train))
```

LinearRegression 함수를 model이라는 이름에 저장

**저장하지 않고 사용하면 학습이 안 됨!!**

```
model = LinearRegression()
```

```
print(type(model))
print(type(LinearRegression()))
print(type(LinearRegression))
```

학습시키고 싶은 데이터셋을 넣고 fit하면 알아서 학습해줌

```
model.fit(x_train, y_train)
```

linear regression으로 만들어진 최적선은 31.04617413 - 0.12402883x 

```
print(model.coef_)
print(model.intercept_)
```

이 모델로 예측한 y(집 가격)

```
y_test_prediction = model.predict(x_test)
```

실제 y값과 예측한 y값의 MSE를 루트 씌움. 

대략 8천달러 정도의 오차가 나왔음을 알 수 있다.

```
mean_squared_error(y_test, y_test_prediction) ** 0.5
```
